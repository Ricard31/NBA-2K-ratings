```python
import numpy as np 
import pandas as pd 
from scipy import stats

# pandas options
pd.set_option('display.max_columns', 60)
pd.set_option('display.float_format', lambda x: '%.2f' % x)

import matplotlib.pyplot as plt
import seaborn as sns

sns.set() # use Seaborn styles
sns.color_palette("Blues")
plt.style.use('fivethirtyeight')

import warnings; warnings.simplefilter('ignore')
```

```python
# Data import
df = pd.read_csv('../2019_ratings_stats.csv')
```

```python
df.head()
```

## EDA

```python
def view_var_importance(df, variable, target):
    if type(variable) == str:
        dataframe = pd.DataFrame({'Mean ' + target: df[target].groupby(df[variable]).mean(), 
                  'Count': df[variable].value_counts() }).sort_values(by=['Mean ' + target, 'Count'], ascending=False)
    else: # variables cut with pd.qcut
        dataframe = pd.DataFrame({'Mean ' + target: df[target].groupby(variable).mean(), 
                  'Count': variable.value_counts() }).sort_values(by=['Mean ' + target, 'Count'], ascending=False)
    print(dataframe)
```

```python
view_var_importance(df, 'Pos', 'rating')
```

```python
view_var_importance(df, 'team', 'rating')
```

```python
vars_list = ['PER', 'WS', 'VORP', 'BPM', 'USG.pct', 'GS', 'PTS']
df2 = df.copy()
# view the effect of those variables in the result
for var in vars_list:
    print('-'*30)
    print(var)
    print('-'*30)
    print('Binning variable', var)
    df2[var] = pd.qcut(df2[var], 5)
    print('-'*30)
    view_var_importance(df2, var, 'rating')
    print('\n')
```

## Highly correlated variables

```python
# Create correlation matrix
corr_matrix = df.corr()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
```
